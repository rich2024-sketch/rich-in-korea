## 11.3 차트 장식을 도와주는 다양한 기법들

$y_1 = 2x$,   
$y_2 = x^2 + 5$   
$y_3 = -x^2 -5$

````python
# import packages
import numpy as np
import matplotlib.pyplot as plt

# x 축 데이터 생겅
x = np.linspace(-20, 20, 41)
# y1, y2, y3 수식 구현
y1 = 2 * x
y2 = x ** 2 + 5
y3 = -x ** 2 - 5
print(x.shape, y3.shape)
# 그래프 그리기
plt.plot(x, y1, 'k_--', label = 'linear')
plt.plot(x, y2, 'g^-', label = 'y2=x^2 + 5')
plt.plot(x, y3, 'b+:', label= 'y3=-x^2 -5')
# 범례
plt.legend()

# 축의 범위 조정하기 # limit the x and y range
plt.ylim([-30, 30])
plt.xlim([-30, 30])

# 축 레이블
plt.xlabel('x')
plt.ylabel('y')

#title
plt.title('linear and quadratics')
plt.show()
````

## 11.4 하나의 차트에 여러 개의 데이터를 그려보자

- linear: $y_1 = x $
- quadratic : $y_2 = x ^ 2$   
- cubic: $y_3 = x ^ 3$

````python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 101)

y1 = x
y2 = x**2
y3 = x**3 

# plt.plot(x, y1, 'k_--')
# plt.plot(x, y2, 'g^-')
# plt.plot(x, y3, 'b+:')

#plotting
plt.plot(x, y1, 'k.-', label = 'linear')
plt.plot(x, y2, 'r^:', label = 'quadratic')
plt.plot(x, y3, 'b*--', label = 'cubic')
plt.ylim([-10, 10])
plt.xlim([-2, 2])
plt.legend()

plt.show()
````

**LAB11-2** 삼각함수의 기본인 사인 그래프 그리기

````python
# import package

# 각도 0-360
theta = np.arange(0, 360)
# 사인 값 만들기
# np.sin(2*np.pi*theta / 360)
y1 = np.sin(np.radians(theta))
y2 = np.cos(np.radians(theta))

# 그리기
plt.plot(theta, y1, label = 'sin')
plt.plot(theta, y2, label = 'cos')
plt.legend()
plt.xlabel('theta')
plt.grid()
plt.show()

````

- 히스토그램 그리는 순서
1) 최대 값과 최소 값을 구한다.
2) 최대 값과 최소 값을 정해진 구간(bin)으로 나눈다.
3) 각 빈에 있는 샘플의 수(도수)를 센다.

두개의 표준 정규 분포 난수를 구하여 히스토그램을 그리자.

````python
import numpy as np
import matplotlib.pyplot as plt

# 표준 정규 분포를 따르는 난수 생성
data1 = np.random.randn(1000)  # 첫 번째 데이터
data2 = np.random.randn(1000)  # 두 번째 데이터

# 히스토그램 그리기
plt.hist(data1, bins=30, alpha=0.5, label='Data 1')  # alpha는 투명도 조절
plt.hist(data2, bins=30, alpha=0.5, label='Data 2')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Two Standard Normal Distributions')
plt.legend()

plt.show()
````
